```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import gensim.downloader as api

# Cargar modelo
model = api.load("word2vec-google-news-300")  # O word2vec-google-news-300
```

```python
vec = lambda word: model[word]
```

```python
y = ("coach", "tutor", "trainer", "educator", "professor", "mentor", "dog", "tree", "cat", "horse")
vec_y= [vec(word) for word in y]
```

```python
diff = vec_y[3] - V_x
e1 = diff /np.linalg.norm(diff)
```

```python
v_tmp = vec_y[7] - V_x
proj_on_e1 = np.dot(v_tmp, e1) * e1
e2 = v_tmp - proj_on_e1
e2 /= np.linalg.norm(e2)
```

```python
def project(v):
    return np.array([np.dot(v, e1), np.dot(v, e2)])
```

```python
points = {
    "teacher": vec("teacher"),
    "coach": vec("coach"),
    "trainer": vec("trainer"),
    "educator": vec("educator"),
    "professor": vec("professor"),
    "mentor": vec("mentor"),
    "dog": vec("dog"),
    "tree": vec("tree"),
    "cat": vec("cat"),
    "horse": vec("horse"),
    "movie": vec("movie")
}
```

```python
plt.figure(figsize=(7, 7))
for word, vector in points.items():
    x, y = project(vector)
    plt.scatter(x, y)
    plt.text(x + 0.01, y + 0.01, word)

plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid()
plt.axis('equal')
plt.show()
```
