#### 1. Limpieza 

Limpiar los documentos de caracteres que no corresponden

```python
doc = datos.iloc[0]['review']
doc
```

```python
doc.replace('<br />', '').replace('>', '').replace('/', '')
```

```python
import re
```

```python
def clean_text(doc):
    return re.sub(pattern=r'<.*?>', repl='', string=doc).replace('.', ' ').replace(',', '').replace('(', '').replace(')', '').replace('"', '').replace("'", '').replace("\x08", '')
```

```python
datos['review'].apply(clean_text)
```

```python
doc.split()
```

#### 2. Normalización

Convertir todos los tokens a minúsculas.

Elimina puntuación y símbolos no alfabéticos.

```python
def normalizacion(doc):
    doc = clean_text(doc)
    doc = doc.lower()
    return doc

# Aplicamos normalización
datos['review_normalizado'] = datos['review'].apply(normalizacion)

print(datos['review_normalizado'].iloc[0])
```

#### 3. Eliminación de Stopwords

Eliminar las palabras vacías (stopwords) usando una lista estándar de la librería _nltk_.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

```

```python
stop_words = set(stopwords.words('english'))

def eliminacion_stopwords(doc):
    tokens = word_tokenize(doc)
    
    filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
    return ' '.join(filtered_tokens)

# Eliminación de stopwords
datos['review_sin_stopwords'] = datos['review_normalizado'].apply(eliminacion_stopwords)
```

```python
print(datos['review_sin_stopwords'].iloc[0])
```

#### 4. Stemming 

Reducir el espacio de palabras

```python
from nltk.stem import porter

stemmer = porter.PorterStemmer()
```

```python
def stemming(doc):
    tokens = doc.split()
    
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

# Stemming
datos['review_stemming'] = datos['review_sin_stopwords'].apply(stemming)
```

```python
print(datos['review_stemming'].iloc[0])
```

#### 5. Verificar la diferencia

Comparar el tamaño del diccionario de términos del corpus antes y después de aplicar el preprocesamiento 

```python
def vocabulario(series):
    all_tokens = []
    for doc in series:
        all_tokens.extend(doc.split())
    return set(all_tokens)


vocabulario_original = vocabulario(datos['review'].apply(lambda x: x.lower()))
vocabulario_normalizado = vocabulario(datos['review_normalizado'])
vocabulario_sin_stopwords = vocabulario(datos['review_sin_stopwords'])
vocabulatio_stemming = vocabulario(datos['review_stemming'])
```

```python
# Resultados
print("="*60)
print("COMPARACIÓN DE VOCABULARIOS")
print("="*60)
print(f"Vocabulario original:        {len(vocabulario_original):,} términos")
print(f"Después de normalización:    {len(vocabulario_normalizado):,} términos")
print(f"Después de quitar stopwords: {len(vocabulario_sin_stopwords):,} términos")
print(f"Después de stemming:         {len(vocabulatio_stemming):,} términos")
print("="*60)
```
